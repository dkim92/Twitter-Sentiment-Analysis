#### The data frames to store the tweets and the sentiment of those tweets are created

#### Since the Search API is used, when a search is performed 25 tweets are returned so a while loop is created to search 100 more times. For each of the tweets the sentiment of the tweet is analyzed using TextBlob to determine if the tweet is positive, negative or neutral. The tweets and the corresponding sentiment is saved into the respective data frames. 

```python
ctr = 0
while ctr <100:
    for status in twitter.search(q="Trump", count= 100, language = 'eng')['statuses']:

        txt = status['text'].encode('utf-8')
        txt = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", txt).split())
        analysis = TextBlob(txt)
        if analysis.sentiment.polarity > 0:
            sent = ['positive']
        elif analysis.sentiment.polarity == 0:
            sent = ['neutral']
        else:
            sent = ['negative']
        tweet = [txt]
        sentiment = sentiment.append(sent)
        tweet_df = tweet_df.append(tweet)
        print (txt, sent)

    ctr = ctr + 1
```
